```python
import pandas as pd

data = [15, 15, 16, 19, 19, 20, 20, 21, 22, 28, 35, 40, 41, 42, 43, 44, 60, 61, 65]

c1 = 16
c2 = 22

c1_list = []
c2_list = []
distance1_list = []
distance2_list = []
nearest_cluster_list = []
new_centroid_list = []

for iteration in range(4):
    c1_list.clear()
    c2_list.clear()
    distance1_list.clear()
    distance2_list.clear()
    nearest_cluster_list.clear()
    new_centroid_list.clear()
    
    for xi in data:
        distance1 = abs(xi - c1)
        distance2 = abs(xi - c2)
        if distance1 < distance2:
            nearest_cluster = 1
        else:
            nearest_cluster = 2
        
        c1_list.append(c1)
        c2_list.append(c2)
        distance1_list.append(distance1)
        distance2_list.append(distance2)
        nearest_cluster_list.append(nearest_cluster)
    
    new_c1 = sum(xi for xi, cluster in zip(data, nearest_cluster_list) if cluster == 1) / nearest_cluster_list.count(1)
    new_c2 = sum(xi for xi, cluster in zip(data, nearest_cluster_list) if cluster == 2) / nearest_cluster_list.count(2)
    
    new_centroid_list.extend([new_c1 if cluster == 1 else new_c2 for cluster in nearest_cluster_list])
    
    df = pd.DataFrame({
        'xi': data,
        'c1': c1_list,
        'c2': c2_list,
        'Distance 1': distance1_list,
        'Distance 2': distance2_list,
        'Nearest Cluster': nearest_cluster_list,
        'New Centroid': new_centroid_list
    })
    
    print(f"Iteration {iteration+1}:")
    print(df.to_string(index=False))
    print()
    
    c1 = new_c1
    c2 = new_c2
```

```python
import pandas as pd

data = [15, 15, 16, 19, 19, 20, 20, 21, 22, 28, 35, 40, 41, 42, 43, 44, 60, 61, 65]

c1 = 16
c2 = 22

kümedeğerilistesi = []

for iteration in range(4):
    kümedeğerilistesi = []
    for xi in data:
        mesafe1 = abs(xi - c1)
        mesafe2 = abs(xi - c2)
        if mesafe1 < mesafe2:
            kümedeğeri = 1
        else:
            kümedeğeri = 2
        
        kümedeğerilistesi.append(kümedeğeri)
    
    count_c1 = kümedeğerilistesi.count(1)
    count_c2 = kümedeğerilistesi.count(2)
    
    if count_c1 > 0:
        yeni_c1 = sum(xi for xi, cluster in zip(data, kümedeğerilistesi) if cluster == 1) / count_c1
    else:
        yeni_c1 = c1
    
    if count_c2 > 0:
        yeni_c2 = sum(xi for xi, cluster in zip(data, kümedeğerilistesi) if cluster == 2) / count_c2
    else:
        yeni_c2 = c2
    
    c1 = yeni_c1
    c2 = yeni_c2
    
result = pd.DataFrame({
    'xi': data,
    'ait olduğu küme': kümedeğerilistesi
})

print(result)

```
